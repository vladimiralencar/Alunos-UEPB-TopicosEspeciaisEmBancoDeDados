```python
import requests

url = 'https://www.saibamais.jor.br/bacurau-sera-mesmo-resistencia/'
page = requests.get(url)
page.text[:300]
```

```python
page.text[:1000]
```

# Pegar 1 texto na web e resumir usando o gensim

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
paragraphs = soup.find_all('p')
paragraphs[:1]
```

```python
text = soup.find_all('p')[0].get_text()
text[:300]
```

## Resumir o texto usando o Gensim

```python
print('Número de parágrafos:', len(soup.find_all('p')))
```

```python
import gensim
num_paragraphs = 3
word_count=1000
print (gensim.summarization.summarize(text, word_count=word_count))
```

```python
# Summary (0.5% of the original content). 
ratio = 0.5
print (gensim.summarization.summarize(text, ratio=ratio))
```

```python
# pip install tika
import tika
tika.initVM()
from tika import parser

url = 'Todas as Coisas Belas - Matthew Quick.pdf'
parsed = parser.from_file(url)
print(parsed["metadata"])
print(parsed["content"])
```

```python
pages = parsed["content"]
len(pages)
```

# Sumarizar o pdf usando o Modelador de Tópicos Gensim

```python
%%time
import gensim

text = pages
# sumariza
num_paragraphs = 3
word_count=4000
#print (gensim.summarization.summarize(text, word_count=word_count, split=True))
print (gensim.summarization.summarize(text))
```

```python
%%time

# Summary (0.5% of the original content). 
ratio = 0.1
print (gensim.summarization.summarize(text, ratio=ratio))
```
