# An√°lise de Sentimentos

```python
def sentiment_analysis(query):
    # creating object of TwitterClient Class
    #api = TwitterClient()
    twitter_api = twitter.Twitter(auth=auth)
    # calling function to get tweets
    #tweets = api.get_tweets(query, count = 1000)
    tweets = twitter_api.search.tweets(query, count = 1000)
    
    print("An√°lise de Sentimentos: ", query)
 
    # picking positive tweets from tweets
    ptweets = [tweet for tweet in tweets if tweet['sentiment'] == 'positive']
    # percentage of positive tweets
    print("Positivo tweets: {0:.0f} %".format(100*len(ptweets)/len(tweets)))
    # picking negative tweets from tweets
    ntweets = [tweet for tweet in tweets if tweet['sentiment'] == 'negative']
    
    # picking neutral tweets from tweets
    neutraltweets = [tweet for tweet in tweets if tweet['sentiment'] == 'neutral']
    
    # percentage of negative tweets
    print("Negativo tweets: {0:.0f} % ".format(100*len(ntweets)/len(tweets)))
    # percentage of neutral tweets
    print("Neutros tweets: {0:.0f} % ".format(100* (len(tweets) - len(ntweets) - len(ptweets)) /len(tweets)))
 
    # printing first 5 positive tweets
    print("\n\nPositivos tweets:")
    for tweet in ptweets[:10]:
        print(tweet['text'])
 
    # printing first 5 negative tweets
    print("\n\nNegativos tweets:")
    for tweet in ntweets[:10]:
        print(tweet['text'])
        
    # printing first 5 neutral tweets
    print("\n\nNeutros tweets:")
    for tweet in neutraltweets[:10]:
        print(tweet['text'])
        
    return ptweets, ntweets, neutraltweets
```

```python
q = 'Anitta'
sentiment_analysis(q)
```
