```python
url = 'covid19-30-janeiro-2022.csv'
url
```

```python
import pandas as pd

df = pd.read_csv(url)
print(len(df))
df.head(8)
```

```python
df.place_type.unique()
```

```python
dfe = df[df.place_type == 'state' ].reset_index(drop=True)
dfe
```

```python
dfo = dfe.sort_values(by='new_confirmed', ascending=False)
dfo
```

```python
dfc = df[df.place_type == 'city' ].reset_index(drop=True)
print(len(dfc))
dfc.head()
```

```python
atributos = dfc.columns
atributos
```

```python
atributos = [ 'city', 'state', 'new_confirmed', 'estimated_population', 'last_available_confirmed' ,
               'last_available_deaths', 'new_deaths' ]
atributos
```

```python
dfc = dfc [ atributos ]
dfc.head()
```

```python
dfco = dfc.sort_values(by='new_confirmed', ascending=False).reset_index(drop=True)
dfco[:20]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.barplot(x='new_confirmed', y='city', data=dfco[:15])
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='new_confirmed', y='city', data=dfco[:15], palette='Reds_r')
```

```python
dfco.head()
```

```python
dfco[dfco.city == 'Campina Grande']
```

```python
dfco[dfco.city == 'Porto Alegre']
```

```python
dfco.head()
```

```python
dfco['tx_transmissibilidade'] = dfco['new_confirmed'] / dfco['estimated_population']
dfco.head()
```

```python
atributos = dfco.columns
atributos = ['city', 'state', 'tx_transmissibilidade', 'new_confirmed', 'estimated_population',
       'last_available_confirmed', 'last_available_deaths', 'new_deaths']
atributos
```

```python
dfco2 = dfco[ atributos ]
dfco2.head()
```

```python
dfco3 = dfco2.sort_values(by='tx_transmissibilidade', ascending=False).reset_index(drop=True)
dfco3[:20]
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='tx_transmissibilidade', y='city', data=dfco3[:15], palette='Reds_r')
```

```python
media = dfco3.tx_transmissibilidade.mean()
media
```

```python
df_acimamedia = dfco3[ dfco3.tx_transmissibilidade > media ]
print(len(df_acimamedia))
df_acimamedia.head()
```
