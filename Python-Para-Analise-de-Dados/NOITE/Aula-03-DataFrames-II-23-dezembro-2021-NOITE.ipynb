```python
url = 'covid19-23-dezembro-2021.csv'
url
```

```python
import pandas as pd

df = pd.read_csv(url)
print(len(df))
df.head(8)
```

```python
df.place_type.unique()
```

```python
dfe = df[df.place_type == 'state' ].reset_index(drop=True)
dfe
```

```python
dfo = dfe.sort_values(by='last_available_confirmed', ascending=False)
dfo
```

```python
dfc = df[df.place_type == 'city' ].reset_index(drop=True)
print(len(dfc))
dfc.head(8)
```

```python
atributos = dfc.columns
atributos
```

```python
atributos = [ 'city', 'state', 'last_available_confirmed', 'new_confirmed',
               'last_available_deaths', 'new_deaths', 'estimated_population']
atributos
```

```python
dfc = dfc [ atributos ]
dfc.head()
```

```python
dfco = dfc.sort_values(by='last_available_confirmed', ascending=False).reset_index(drop=True)
dfco[:20]
```

```python
atributos = dfco.columns
atributos
```

```python
atributos = [ 'state', 'last_available_confirmed', 'new_confirmed',
       'last_available_deaths', 'new_deaths', 'estimated_population']
atributos
```

## Eliminar as cidades duplicadas

```python
print(len(dfc))
dfc2 = dfc.groupby('city')[atributos].max().reset_index()
print(len(dfc2))
dfc2.head()
```

```python
dfco = dfc2.sort_values(by='last_available_confirmed', ascending=False).reset_index(drop=True)
dfco[:20]
```

```python
dfco = dfc2.sort_values(by='new_confirmed', ascending=False).reset_index(drop=True)
dfco[:20]
```

```python
cidades = list(dfc.city.drop_duplicates().values)
cidades
```

```python
dfco.head()
```

```python
print(len(dfc))
dfc2 = dfc.loc [ : ,   (cidades == dfc.city) ]
print(len(dfc2))
```

```python
mapa = df.merge(dfn, left_on='sigla', right_on='state')
mapa
```

# Mapa Temático ou Choropleth map
