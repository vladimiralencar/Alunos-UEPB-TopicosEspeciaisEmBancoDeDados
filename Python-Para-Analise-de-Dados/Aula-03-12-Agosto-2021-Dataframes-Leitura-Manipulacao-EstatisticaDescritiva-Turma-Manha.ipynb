```python
ds = dfe.sort_values(by='last_available_confirmed', ascending=False)
ds.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))

sns.barplot(x="last_available_confirmed", y="state", data=ds)
```
