```python
sul = [ 'RS', 'SC', 'PR']
sudeste = [ 'MG', 'SP', 'ES', 'RJ']
norte = [ 'AM', 'RR', 'RO', 'PA', 'AC', 'AP', 'TO']
nordeste = ['BA', 'SE', 'AL', 'PE', 'PB', 'RN', 'MA', 'PI', 'CE' ]
centro_oeste =  ['DF', 'GO', 'MT', 'MS' ] 
```

```python
df['regioes'] = df.apply(lambda x: 'Sudeste' if x['ESTADO'] in sudeste else x['regioes'], axis=1)
df['regioes'] = df.apply(lambda x: 'Sul' if x['ESTADO'] in sul else x['regioes'], axis=1)
df['regioes'] = df.apply(lambda x: 'Norte' if x['ESTADO'] in norte else x['regioes'], axis=1)
df['regioes'] = df.apply(lambda x: 'Nordeste' if x['ESTADO'] in nordeste else x['regioes'], axis=1)
df['regioes'] = df.apply(lambda x: 'Centro-Oeste' if x['ESTADO'] in centro_oeste else x['regioes'], axis=1)
df.head()
```
