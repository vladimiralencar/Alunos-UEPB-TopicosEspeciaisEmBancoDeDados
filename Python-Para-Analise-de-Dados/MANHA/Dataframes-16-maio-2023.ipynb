```python
# Jun√ß√£o
d = df.merge(dfm, left_on='sigla', right_on='STATE')
d.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# palettes
pal = ["Reds", 'OrRd', 'cubehelix', 'Greens', 'winter', 'Greens', "RdBu_r", 'Blues', 'OrRd' ]

cmap = pal[0]


fig, ax = plt.subplots(figsize=df.head()(14, 10))

d.plot(column='GDP', figsize=(14,10), cmap=cmap, legend=True, ax=ax)

ax.set(title='Brasil PIB')

ax.set_axis_off()
plt.show()
```
